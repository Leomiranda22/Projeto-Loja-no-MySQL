```python
from faker import Faker
import random

fake = Faker()

def gerar_venda():
    cliente_id = random.randint(1, 100)  
    data_venda = fake.date_time_between(start_date='-1y', end_date='now').strftime('%Y-%m-%d %H:%M:%S')
    total = round(random.uniform(50.00, 1000.00), 2)
    return (cliente_id, data_venda, total)


vendas = [gerar_venda() for _ in range(400)]

for venda in vendas:
    print(f"INSERT INTO lojaLMS.Vendas (ClienteID, DataVenda, Total) VALUES ({venda[0]}, '{venda[1]}', {venda[2]});")
```

```python
from faker import Faker
import random

fake = Faker()

produtos_existentes = list(range(1, 12))

def gerar_item_vendido():
    venda_id = random.randint(1, 400)  
    produto_id = random.choice(produtos_existentes)  
    quantidade = random.randint(1, 10)  
    preco_unitario = round(random.uniform(10.00, 200.00), 2) 
    return (venda_id, produto_id, quantidade, preco_unitario)

itens_vendidos = [gerar_item_vendido() for _ in range(500)]

for item in itens_vendidos:
    print(f"INSERT INTO lojaLMS.ItensVendidos (VendaID, ProdutoID, Quantidade, PrecoUnitario) VALUES ({item[0]}, {item[1]}, {item[2]}, {item[3]});")
```

```python
from faker import Faker
import random

fake = Faker()

def estado_civil():
    return random.choice(['Solteiro', 'Casado'])

for i in range(10):
    cliente = fake.name()
    estado = fake.state()
    civil = estado_civil()
    print (f"INSERT INTO (cliente, estado, estadocivil) VALUES ('{cliente}'), '{estado}', '{civil}');")
```

```python
class Carro:
    def __init__(self, marca, modelo, cor):
        self.marca = marca
        self.modelo = modelo
        self.cor = cor

    def andar(self):
        print (f'Estou andando')

    def parar(self):
        print(f'Estou parando para exibir as config')

    def exibir_config(self):
        print(self.marca, self.modelo, self.cor)


carro1 = Carro('Gol', 'Golfão', 'Azul')
carro1 = andar()
carro1 = parar()
carro1 = exibir_config()
```
